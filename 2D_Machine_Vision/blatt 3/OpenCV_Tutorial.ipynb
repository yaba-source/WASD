## OpenCV in Python importieren
In Python muss nun OpenCV import werden (Paket <code>cv2</code>), ebenso <code>matplotlib</code> für die Visualisierung und <code>numpy</code> für den Zugriff auf Pixel:

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

Bilder können mit <code>cv.imread()</code> gelesen werden und anschließend dargestellt werden.

```python
img = cv.imread("/workspaces/WASD/2D_Machine_Vision/robot_gray.png")

plt.imshow(img)
plt.axis('off')
plt.show()
```

Bei OpenCV Bildern handelt es sich um mehrdimensionale numpy Arrays:

```python
print(type(img), img.dtype, img.shape, img[0,0,:])
```

Diese können direkt modifiziert werden:

```python
new_img = np.copy(img)
new_img[0:100,0:100,:] = 0

plt.imshow(new_img)
plt.axis('off')
plt.show()
```

OpenCV bietet diverse Funktionen, um Bilder zu modifizieren. Hier z.B. die Umwandlung in ein grauwert Bild:

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap="gray")
plt.axis('off')
plt.show()
```

OpenCV kann auch direkt von einer Kamera oder aus Videos Bilder aufnehmen. Zum Beispiel durch
* eine IP Kamera: <code>cv.VideoCapture("http://\<IP-Adresse\>")</code>, oder
* eine direkt angeschlossene Kamera (unter Linux z.B. unter "/dev/video0"): <code>cv2.VideoCapture(0)</code>, oder
* aus einem Video: <code>cv2.VideoCapture('video.mp4')</code>

```python
cam = cv.VideoCapture("http://host.docker.internal:56000/jpeg")

success, frame = cam.read()
if success:
    plt.imshow(cv.cvtColor(frame, cv.COLOR_RGB2BGR))
    plt.axis('off')
    plt.show()
else:
    print("Failed")
```
